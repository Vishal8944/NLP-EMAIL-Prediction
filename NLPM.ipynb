```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report
```

```python
df=pd.read_csv(r"C:\Users\Vishal S\OneDrive\Desktop\email.csv",encoding="utf-8")
df
```

```python
nltk.download('stopwords')
```

```python
df['Label'] = df['Label'].map({'Spam': 1, 'Non-Spam': 0})
```

```python
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = text.strip()  # Remove extra spaces
    text = " ".join(word for word in text.split() if word not in stopwords.words('english'))  # Remove stopwords
    return text
```

```python
df['Cleaned_Message'] = df['Message_body'].apply(clean_text)
```

```python
X=df['Cleaned_Message']
Y=df['Label']
```

```python
X
```

```python
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.25, random_state=12)
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
def predict_sms(text):
    cleaned_text = clean_text(text)
    prediction = model.predict([cleaned_text])[0]
    return "Spam" if prediction == 1 else "Non-Spam"
```

```python
print(predict_sms("Hey, are you coming to the party tonight?"))
```
